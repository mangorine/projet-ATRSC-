## Notations

On définit quelques termes et on les note comme ceci:
- $L_j$: Nombre moyen de produits dans la machine $j$ (En vrai c'est plutôt nombre tout court mais on prendra l'espérance après ça change rien donc autant le définir comme ça et éviter d'écrire "espérance" à chaque fois)
- $\lambda_j$: $\lambda$ associé à la machine $j$ (c'est à dire l'intensité d'arrivée sur la machine $j$)
- $W_j$: Temps d'attente moyen sur la machine $j$ (même chose que pour le $L_j$)
- $\lambda_t$: $\lambda$ associé au produit $t$ (le type $t$)
- $L$: Nombre moyen total
- $W$: Temps d'attente moyen total
- $S_{t,j}$: Temps de service moyen du produit $t$ à la machine $j$ (même chose, on prend l'espérance)
 
## Première estimation de la borne inférieure

## Le piège du modèle M/G/1

Au départ, je voulais modéliser chaque employé par une file d'attente et dire que le temps d'attente moyen réel serait plus élevé que le temps d'attente pour une M/G/1 pour par la suite utiliser Pollaczek-Khinchine.
On pourrait être tenté de faire cela pour calculer l'attente due à la congestion.
Mais attention, pour nos instances (notamment l'employé goulot de l'instance 1), les arrivées ne sont plus poissonniennes. De fait, un produit arrive sur une machine après avoir subi un temps de service déjà effectué sur une autre machine en amont.

## Autre méthode: employé goulot

On procède alors à considérer l'employé qui serait le plus chargé (celui qui a le $\rho$ le plus élevé) et à dire que:

$$
W \geq W_{opt} + W_{goulot}
$$

## L'effet de lissage et le modèle IFR/G/1

Le fait que le temps de traitement précédent suive une loi uniforme en fait "lisse" le flux de produits. La variance des inter-arrivées diminue drastiquement par rapport à un processus de Poisson. Mathématiquement, les temps d'inter-arrivée suivent une distribution de type **IFR** (Increasing Failure Rate, j'ai trouvé un article sur ça donc on va utiliser ce modèle).
En clair : plus on attend un produit depuis la dernière arrivée, plus il a de chances d'arriver à l'instant suivant (contrairement à Poisson qui est sans mémoire).

Nous sommes donc face à une file **IFR/G/1**. Utiliser la formule classique M/G/1 surestimerait l'attente et fausserait notre borne inférieure (on risquerait de dépasser le vrai temps d'attente).

## La borne de Marshall-Stoyan

Puisque nous avons des arrivées régularisées (IFR), nous pouvons utiliser l'inégalité de Marshall-Stoyan (ou la borne de Mori) qui nous donne une estimation de l'attente minimale dans la file ($W_q$) pour l'employé goulot :

$$
W_{q, goulot} \geq \frac{\lambda_{goulot} \text{Var}(S_{goulot})}{2(1-\rho_{goulot})} - \frac{1}{2}\mathbb{E}[S_{goulot}]
$$

Notre véritable borne inférieure heuristique pour le temps de séjour total devient donc la somme des temps de service optimaux, à laquelle on ajoute cette attente inévitable au goulot due à la variabilité de la loi uniforme :

$$
W \geq W_{opt} + \max \left( 0, \frac{\lambda_{goulot} \text{Var}(S_{goulot})}{2(1-\rho_{goulot})} - \frac{1}{2}\mathbb{E}[S_{goulot}] \right)
$$
