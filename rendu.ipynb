## Notations

On définit quelques termes et on les note comme ceci:
- $L_j$: Nombre moyen de produits dans la machine $j$ (En vrai c'est plutôt nombre tout court mais on prendre l'espérance après ça change rien donc autant le définir comme ça et éviter d'écrire espérance à chaque fois)
- $\lambda_j$: $\lambda$ associé à la machine $j$ (c'est à dire l'intensité d'arrivée sur la machine $j$)
- $W_j$: Temps d'attente moyen sur la machine $j$ (même chose que pour le $L_j$)
- $\lambda_t$: $\lambda$ associé au produit $t$ (le type $t$)
- $L$: Nombre moyen total
- $W$: Temps d'attente moyen total
- $S_{t,j}$: Temps de service moyen du produit $t$ à la machine $j$ (même chose, on prend l'espérance)

## Formule de Pollaczek-Khinchine

On rappelle la formule de Pollaczek-Khinchine (PK) qui explicite le temps d'attente moyen pour une M/G/1:
$$
W = \underbrace{\frac{\sum\limits_{t \in Produits}\mathbb{E}[S_t^2]}{2(1-\rho)}}_{\text{Attente dans la file}} + \underbrace{\frac{\rho}{\lambda}}_{\text{Temps de service}}
$$
