- Une borne absolue est le temps de traitement physique en supposant une attente nulle: 

$$ \sum\limits_{k=1}^4 {\lambda_k / \lambda_{total}}{\mathbb{E[S_k]}} = 2.1 \quad \text{unite de temps}

    - $\rho_{M2}  \approx 0.31$
    - $\rho_{M3}  \approx 0.47$
    - $\rho_{M4}  \approx 0.32$

    - $\rho_{M8}  \approx 0.37$
    - Charge globale de l'atelier : $\rho_{global} = \sum \rho_j \approx 3.05$ unites de travail / u.t.
- **I1** : Cette instance partitionne strictement l'atelier avec $q=4$ employés. Chaque employe gere un binome de machines fixe. Les charges de travail resultantes par employe sont :
    - Employé 1 (M1, M2) : $\approx 0.80$ % d'occupation
    - Employé 2 (M3, M4) : $\approx 0.78$ % d'occupation
    - Employé 3 (M5, M6) : $\approx 0.78$ % d'occupation
    - Employé 4 (M7, M8) : $\approx 0.68$ % d'occupation
Bien que stable ($\rho < 1$), le systeme subit une forte congestion, notamment sur le premier binome. La file d'attente moyenne augmente de facon non-lineaire avec la charge, rendant la borne de I1 > 2.09.

- **I2** : Avec $q=4$ employes et des competences croisees, la charge globale ($\rho_{global} \approx 3.05$) est répartie de maniere plus flexible. La charge moyenne par employe reste a $\rho \approx 0.76$, mais la polyvalence permet de "lisser" les pics d'arrivee. La borne est > 2.09, mais plus basse que celle de I1.

- **I3** : Le systeme dispose de $q=6$ employe avec une matrice de qualifications creuse. La charge moyenne tombe a $\rho \approx 0.51$. Les ressource sont plus abondantes, limitant les files d'attente. La borne est proche de 2.09.

- **I4** : Avec $q=6$ employe et une hyper-polyvalence, la flexibilite est maximale. Un produit arrivant sur une machine trouvera presque instantanement un employe qualifie disponible en salle d'attente. L'attente tend vers zero, et la borne inférieure est quasiment egale à 2.09.
