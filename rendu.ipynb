# Question 1


Pour qu'un tel système de files d'attente atteigne un régime permanent, deux conditions heuristiques doivent être satisfaites : la **faisabilité** (les ressources sont suffisantes pour couvrir les tâches) et la **stationnarité** (la capacité de traitement est supérieure à la demande).

Il est nécessaire que chaque machine $j$ puisse être opérée par au moins un employé $i$. Mathématiquement, cela se traduit par le fait qu'aucune colonne de la matrice des qualifications $Q$ ne soit nulle :
$$
\forall j \in \llbracket 1,m \rrbracket, \quad \sum\limits_{i=1}^q Q_{ij} \geq 1 \quad \text{(Condition de faisabilité)}\\
$$

Pour éviter que les produits ne s'accumulent indéfiniment devant une machine, le taux d'occupation $\rho_j$ de chaque machine $j$ doit être strictement inférieur à 1. La charge $\rho_j$ est la somme des produits des intensités d'arrivée $\lambda_i$ par l'espérance du temps de service $E[S]_{i,j}$ pour tous les types de produits $i$ passant par cette machine :

$$
\forall j \in \llbracket 1,m \rrbracket, \quad \sum\limits_{i=1}^k {\lambda_i}{\mathbb{E[S]}_j^{(i)}} < 1 \quad \text{(Condition de stationarité)}
$$

**AN**: La condition de faisabilite est facilement verifiee, suffit de regarder des matrices.

Pour la condition de stationarité:

- $\rho_{M2}  \approx 0.31$

- $\rho_{M4}  \approx 0.43$
- $\rho_{M5}  \approx 0.40$
- $\rho_{M6}  \approx 0.25$

- $\rho_{M8}  \approx 0.37$

# Question 2

Analysons les différents indicateurs proposés :
- **(i) Le temps de séjour moyen** : ce critère est très pertinent. Le temps de séjour global d’un produit est la somme de ses temps de traitement sur les machines et de ses temps d’attente dans les files. Les temps de traitement étant incompressibles, l’objectif d’ALGO est précisément d’optimiser l’affectation des employés afin de fluidifier le parcours et de minimiser ces temps d’attente. C’est donc un excellent indicateur de performance.

- **(ii) Le nombre moyen de produits traités par unité de temps**, c’est-à-dire le débit : cet indicateur n’est pas pertinent pour différencier les algorithmes dans ce contexte. En effet, puisque le système est en régime permanent, le principe de conservation des flux implique que le taux de sortie est strictement égal au taux d’arrivée des produits. L’intensité des arrivées étant dictée de manière exogène par un processus de Poisson, le débit final sera identique pour n’importe quel algorithme maintenant le système stable.

- **(iii) Le nombre moyen de produits présents dans l’atelier à tout instant** : bien que cet indicateur reflète l’encombrement de l’atelier, il est redondant avec l’indicateur (i). D’après le théorème de Little, qui s’écrit $L = \lambda W$, le nombre moyen de clients dans le système ($L$) est proportionnel au temps de séjour moyen ($W$), car le taux d’arrivée moyen ($\lambda$) est constant. Par conséquent, minimiser le nombre de produits présents revient mathématiquement au même que minimiser le temps de séjour moyen.

## Notations

On définit quelques termes et on les note comme ceci:
- $L_j$: Nombre moyen de produits dans la machine $j$ (En vrai c'est plutôt nombre tout court mais on prendra l'espérance après ça change rien donc autant le définir comme ça et éviter d'écrire "espérance" à chaque fois)
- $\lambda_j$: $\lambda$ associé à la machine $j$ (c'est à dire l'intensité d'arrivée sur la machine $j$)
- $W_j$: Temps d'attente moyen sur la machine $j$ (même chose que pour le $L_j$)
- $\lambda_t$: $\lambda$ associé au produit $t$ (le type $t$)
- $L$: Nombre moyen total
- $W$: Temps d'attente moyen total
- $S_{t,j}$: Temps de service moyen du produit $t$ à la machine $j$ (même chose, on prend l'espérance)

## Le piège du modèle M/G/1

Au départ, je voulais modéliser chaque employé par une file d'attente et dire que le temps d'attente moyen réel serait plus élevé que le temps d'attente pour une M/G/1 pour par la suite utiliser Pollaczek-Khinchine.
On pourrait être tenté de faire cela pour calculer l'attente due à la congestion.
Mais attention, pour nos instances (notamment l'employé goulot de l'instance 1), les arrivées ne sont plus poissonniennes. De fait, un produit arrive sur une machine après avoir subi un temps de service déjà effectué sur une autre machine en amont.

## Autre méthode: employé goulot

On procède alors à considérer l'employé qui serait le plus chargé (celui qui a le $\rho$ le plus élevé) et à dire que:

$$
W \geq W_{opt} + W_{goulot}
$$

## L'effet de lissage et le modèle IFR/G/1

Le fait que le temps de traitement précédent suive une loi uniforme en fait "lisse" le flux de produits. La variance des inter-arrivées diminue drastiquement par rapport à un processus de Poisson. Mathématiquement, les temps d'inter-arrivée suivent une distribution de type **IFR** (Increasing Failure Rate, j'ai trouvé un article sur ça donc on va utiliser ce modèle).
En clair : plus on attend un produit depuis la dernière arrivée, plus il a de chances d'arriver à l'instant suivant (contrairement à Poisson qui est sans mémoire).

Nous sommes donc face à une file **IFR/G/1**. Utiliser la formule classique M/G/1 surestimerait l'attente et fausserait notre borne inférieure (on risquerait de dépasser le vrai temps d'attente).

## La borne de Marshall-Stoyan

Puisque nous avons des arrivées régularisées (IFR), nous pouvons utiliser l'inégalité de Marshall-Stoyan (ou la borne de Mori) qui nous donne une estimation de l'attente minimale dans la file ($W_q$) pour l'employé goulot :

$$
W_{q, goulot} \geq \frac{\lambda_{goulot} \text{Var}(S_{goulot})}{2(1-\rho_{goulot})} - \frac{1}{2}\mathbb{E}[S_{goulot}]
$$

Notre véritable borne inférieure heuristique pour le temps de séjour total devient donc la somme des temps de service optimaux, à laquelle on ajoute cette attente inévitable au goulot due à la variabilité de la loi uniforme :

$$
W \geq W_{opt} + \max \left( 0, \frac{\lambda_{goulot} \text{Var}(S_{goulot})}{2(1-\rho_{goulot})} - \frac{1}{2}\mathbb{E}[S_{goulot}] \right)
$$
